```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tqdm.keras import TqdmCallback  # Progress bar for Keras training
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

data_dir = r"/content/drive/MyDrive/7th Sem/DL/train"
img_size = (128, 128)
batch_size = 32

# Data preparation
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42
)

val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=42
)

# Define Model 1 (Dense-only network)
def create_model():
    model = Sequential([
        Flatten(input_shape=(img_size[0], img_size[1], 3)),
        Dense(128, activation='relu'),
        Dense(256, activation='relu'),
        Dense(256, activation='relu'),
        Dense(128, activation='relu'),
        Dense(train_generator.num_classes, activation='softmax')
    ])
    return model

# Different learning rates to test
learning_rates = [0.1, 0.01, 0.001, 0.0001]

# Dictionary to store results
results = {}

# Train model with different learning rates
for lr in learning_rates:
    print(f"\nTraining Model with learning rate {lr}")

    # Create model
    model = create_model()

    # Compile with specific learning rate
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=optimizer,
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])

    # Train with progress bar
    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=10,
        verbose=0,  # Disable default progress output
        callbacks=[TqdmCallback(verbose=1)]  # Enable tqdm progress bar
    )

    # Store results
    key = f"Learning Rate = {lr}"
    results[key] = history.history

    # Evaluate the model on the validation set
    y_pred_probs = model.predict(val_generator)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = val_generator.classes[val_generator.index_array] # Get true labels from the generator

    # Get the unique labels present in the true and predicted labels
    unique_labels = np.unique(np.concatenate((y_true, y_pred)))
    target_names = [list(val_generator.class_indices.keys())[i] for i in unique_labels]

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=target_names, labels=unique_labels))


    # Calculate and display confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=unique_labels)
    print("\nConfusion Matrix:")
    print(cm)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(f'Confusion Matrix Heatmap (Learning Rate = {lr})')
    plt.show()

# Plotting results
plt.figure(figsize=(15, 5))

# Plot accuracy comparison
plt.subplot(1, 2, 1)
for model_name, history in results.items():
    plt.plot(history['val_accuracy'], label=model_name)
plt.title('Validation Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Plot loss comparison
plt.subplot(1, 2, 2)
for model_name, history in results.items():
    plt.plot(history['val_loss'], label=model_name)
plt.title('Validation Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```
